# Excercise 5

### 1. Create a new column with the following condition: a) >10    : Long Trip b) 5-10   : Medium Trip c) <5     : Short Trip

```python
conditions = [
    (df['MILES*'] < 5),
    (df['MILES*'] >= 5) & (df['MILES*'] <= 10),
    (df['MILES*'] > 10)
]

category = ['Short Trip', 'Medium Trip', 'Long Trip']

df['MILES CAT'] = np.select(conditions, category)

df
```
