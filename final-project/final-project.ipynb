# Solve this Missing Values Problem

### a. Drop column which has missing value and unlogic values such as host_id, name, etc

```python
df.columns
```

```python
drop_columns = ['id', 'name', 'host_id', 'host_name']
df.drop(drop_columns, axis=1, inplace=True)
```

### b. Fill the missing values in column last_review using modes

```python
df['last_review'] = df['last_review'].fillna(df['last_review'].mode()[0])
```

### c. Fill the missing values in column review_per_month using mean

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(df['reviews_per_month'].mean())
```

# Create Category based on Price: Cheap, Medium, and Expensive

```python
print(df['price'].min())
print(df['price'].mean())
print(df['price'].max())
```

```python
conditions =[
    (df['price'] < 100),
    (df['price'] >= 100) & (df['price'] <= 200),
    (df['price'] > 200)
]

category = ['Cheap', 'Medium', "Expensive"]

df['price-category'] = np.select(conditions, category)
```

# Filter All Data with last_review > 2019
